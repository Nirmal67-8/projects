```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load your dataset here (replace with your actual data)
def load_data():
    df = pd.read_csv('/content/drive/MyDrive/PROJECT_DATA/Hyderabad Region new.csv', index_col=False)
    df.drop(['Sr.No'], inplace=True, axis=1)
    return df



# Define a function for the homepage
def home():
    st.title('Home')
    # Display a gif from a file
    gif_file_path = "/content/drive/MyDrive/PROJECT_DATA/Untitled design.gif"
    st.image(gif_file_path, use_column_width=True)
    st.write("Welcome to the Home section of our Exploratory Data Analysis (EDA) app.")
    st.write("Here, you can explore and analyze data related to the Hyderabad Region. There are 8 Districts:")
    st.write("1. Badin")
    st.write("2. Dadu")
    st.write("3. Hyderabad")
    st.write("4. Jamshoro")
    st.write("5. Matiari")
    st.write("6. Sujawal")
    st.write("7. Tando Allah Yar")
    st.write("8. Tando Muhammad Khan")
    st.write("DATA SOURCE: Director schools education hyderabad")


# Define a function for the "Why Focus on Enrollment Rate" section
def focus_on_enrollment():
    st.title('Why Focus on Enrollment Rate')

    # Checkbox to display content
    display_content = st.checkbox('Targeted column ')

    if display_content:
        long_text = """
        TARGET COLUMN IS TOTAL ENROLLMENT:
          Problem statment or the factors affected by enrollment rate:

        Financial Strain:
        One of the immediate consequences of low enrollment is financial strain on schools. Public schools often
        receive funding based on the number of students they enroll. When enrollment is low, schools may struggle
        to cover their operating costs and provide necessary resources for students. This financial strain can lead
        to a decline in the quality of education and overall school performance.

        Reduced Program Offerings:
        Low enrollment can result in reduced program offerings within schools. With limited funds, schools may have
        to cut back on essential programs, extracurricular activities, and elective courses. As a result, students
        may have access to a narrower range of educational opportunities, limiting their overall growth and development.

        Limited Extracurricular Activities:
        Extracurricular activities, such as sports, clubs, and arts programs, play a vital role in a student's education.
         However, low enrollment can lead to a reduction in these activities, affecting students' overall educational
        experience and personal development. Fewer extracurricular options may limit students' opportunities to explore their interests and talents.

        Impact on Communities:
        Public schools often serve as central hubs within communities. When enrollment is critically low, schools may face
        the possibility of closure. The closure of a school due to low enrollment can have significant social and economic
         consequences. Communities lose a valuable resource, which can impact property values and community cohesion.

        Access to Education:
        Low enrollment rates can limit access to education for residents in a particular area. Students may be required to
        travel longer distances to attend schools with higher enrollment, leading to transportation challenges and increased
         commuting times. This can be particularly burdensome for students and families.

        Economic Development:
        Access to quality education is a critical factor for businesses and families considering relocation. Low enrollment
        rates can deter businesses and families from moving to an area, potentially impacting economic development and job
        opportunities in the long run. A well-educated workforce is essential for economic growth.

        To address the impact of low enrollment rates in public schools, education authorities and local governments
         must consider various strategies. These strategies may include consolidating schools, targeted marketing and
         outreach to boost enrollment, optimizing resource allocation, and exploring innovative educational models.
         These efforts aim to ensure that students receive a quality education while addressing the financial and
         community challenges associated with low enrollment.
        """
        st.text(long_text)


def story():
    st.title('Storytelling for the Project')
    long_text = """
    Introduction
    I come from a village where, initially, only primary education was provided. Although
    there were buildings for middle and secondary education,the lack of teaching staff, basic facilities,
     books, furniture, and financial support meant that they remained unused. For years,
    we faced numerous struggles. It was not until around 2007 that we began to see a change in our village's educational landscape.

    Challenges Faced
    During those early years, the primary challenge was the low enrollment rate.
    Students were hesitant to attend school due to the lack of resources
    and support. This posed a significant obstacle to our educational aspirations, particularly for girls.
    In rural areas, the concept of female education was nearly nonexistent.
    Girls were not allowed to leave the village for education.

    Middle School Opportunity
    However, after considerable effort and advocacy, we managed to secure the establishment of a middle school in our village.
    This marked a turning point, especially for female students. The increased enrollment led to a higher ratio of girls attending school.

    Secondary Education
    As the years passed, our village continued to progress; the middle school eventually expanded to include secondary education.
    With two primary schools, two middle schools, and secondary schools, we finally had the educational infrastructure we had longed for,
    although the facilities remained basic.

    Improved Literacy Rate
    The impact of these changes was remarkable. Our village's literacy rate began to rise steadily, surpassing that of neighboring villages.
    Students from grade 5 and 8 were achieving impressive results, and our village became known for its dedication to education.

    Scholarships and Higher Education
    One of the most encouraging developments was the success of our students in various scholarship programs. Many students from
    our village secured scholarships to prestigious institutions like Pak-Turk international school and college Karachi,
    AQ Khan School & College in Islamabad and IBA public schools, often with full scholarships. Our village became a hub for aspiring medical students,
    many of whom are now pursuing MBBS and nursing degrees.

    Attracting Students from Outside
    The transformation did not stop there. Today, students from cities and neighboring areas are coming to our village for education.
    Our educational institutions, despite their humble beginnings, have become beacons of hope and excellence.
    This story showcases the incredible journey of a village that overcame challenges, transformed its education system,
    and became a source of inspiration and opportunity for its residents and beyond.

    Winning the prestigious International School Award from the British Council was a significant milestone for our educational institution.
    This recognition not only underscored our commitment to providing quality education but also highlighted our dedication to fostering
    global citizenship among our students.

    In addition to this remarkable achievement, our students have consistently excelled in science fairs,
    showcasing their ingenuity and passion for scientific inquiry. These victories have not only brought pride to our school but
    have also served as a testament to the quality of education and mentorship our students receive. Furthermore, our commitment to advancing technology
    and computer education has been unwavering. Through diligent efforts, we secured funding for laptops, empowering our students with
    modern tools to enhance their digital literacy. This initiative represents a critical step forward in preparing our students for the demands
    of the digital age and the ever-evolving technological landscape.
    """

    st.text(long_text)

# Define a function for the "Insights" section
def insights():
    st.title('Data Insights')

    long_text = """
        Data Insights
        FACTOR AFFECTING TOTAL ENROLLMENT:

    In our exploration of the dataset, we focused on understanding the factors that influence the "Total Enrollment" in educational
     institutions across the Hyderabad Region. This analysis revealed key insights into the relationship between enrollment
     and several important variables. Let's delve into these findings:

        Highly Correlated Factors

     1. Total Teaching Staff:
       - There is a strong positive correlation between the number of teaching staff and total enrollment.
       - Increasing the teaching staff can lead to higher enrollment, possibly due to improved
       teacher-to-student ratios and enhanced educational experiences.

     2. Number of Male Teachers and Female Teachers:
       - Both the number of male and female teachers show a significant positive correlation with total enrollment.
       - A balanced and adequate teaching staff comprising both genders can create a
        more inclusive and conducive learning environment.

     3. Girls Enrollment and Boys Enrollment:
       - Unsurprisingly, both girls' and boys' enrollment positively impact total enrollment.
       - Encouraging enrollment for both genders is crucial for overall educational development.

     4. Total Classrooms and Rooms:
       - The availability of classrooms and additional rooms exhibits a positive correlation with total enrollment.
       - Adequate infrastructure ensures that educational institutions can accommodate more students comfortably.

         Moderately Correlated Factors

     1. Primary Classrooms:
       - Primary classroom availability displays a moderate positive correlation with total enrollment.
       - Enhancing primary classroom facilities can contribute to increased enrollment,
        especially at the foundational level.

      2. Female Teachers:
       - The presence of female teachers shows a moderate positive correlation with total enrollment.
       - Having a diverse teaching staff may encourage more female students to enroll.

    These insights highlight the significance of investing in educational resources such as
    teaching staff, classrooms, and gender-diverse teachers. By addressing these factors and fostering a conducive learning environment,
     educational institutions can potentially boost enrollment rates and contribute to the educational development of the Hyderabad Region.

    Remember, these correlations provide valuable insights, but a comprehensive strategy for educational improvement
     should consider various other factors, including socioeconomic conditions and community engagement.
    """

    st.text(long_text)





# Define a function to show raw data
def show_raw_data(df):
    st.title('Raw Data')
    if st.checkbox('Show Raw Data'):
        st.write(df)

# Define a function to show numerical columns
def show_numerical_columns(df):
    st.title('Numerical Columns')
    if st.checkbox('Show Numerical Columns'):
        numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
        st.write(numerical_columns)

# Define a function to show categorical columns
def show_categorical_columns(df):
    st.title('Categorical Columns')
    if st.checkbox('Show Categorical Columns'):
        categorical_columns = df.select_dtypes(include=['object']).drop(columns=['School Name', 'District']).columns
        st.write(categorical_columns)


# Define a function to show summary statistics
def show_summary(df):
    st.title('Summary Statistics')
    show_summary_checkbox = st.checkbox('Show Summary/Describe')
    show_numerical_columns_checkbox = st.checkbox('Show Numerical Columns', key='show_numerical_columns')

    if show_summary_checkbox:
        if not show_numerical_columns_checkbox:
            st.write("Summary Statistics:")
            st.write(df.describe())
        else:
            st.write("Summary Statistics (Numerical Columns Only):")
            st.write(df.describe(include=['int64', 'float64']))



# Define a function to show correlation heatmap and table
def show_correlation_heatmap_and_table(df):
    st.title('Correlation Heatmap and Correlation Table')
    display_correlation = st.checkbox('Display Correlation Heatmap and Table')

    if display_correlation:
        numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
        correlation_matrix = df[numeric_columns].corr()
        fig = px.imshow(correlation_matrix,
                        x=numeric_columns,
                        y=numeric_columns,
                        color_continuous_scale='Viridis',
                        labels=dict(color='Correlation'),
                        title='Correlation Heatmap')

        fig.update_xaxes(side='top')
        fig.update_layout(height=800, width=800)

        # Add hovertemplate to display correlation values and highlight highly correlated values
        hovertemplate = '%{x} and %{y}<br>Correlation: %{z:.2f}'
        fig.update_traces(hovertemplate=hovertemplate, hoverongaps=False)

        # Highlight highly correlated values (>= 0.7) in red and moderately correlated values (>= 0.4) in blue
        threshold_high = 0.7
        threshold_moderate = 0.4
        for i in range(len(numeric_columns)):
            for j in range(len(numeric_columns)):
                correlation_value = correlation_matrix.iloc[i, j]
                if i != j:
                    if abs(correlation_value) >= threshold_high:
                        color = 'red'
                    elif abs(correlation_value) >= threshold_moderate:
                        color = 'blue'
                    else:
                        color = 'white'

                    fig.add_shape(type='rect',
                                  x0=i - 0.5,
                                  x1=i + 0.5,
                                  y0=j - 0.5,
                                  y1=j + 0.5,
                                  fillcolor=color,
                                  opacity=0.5,
                                  line=dict(width=0),
                                  xref='x',
                                  yref='y')

        st.plotly_chart(fig, use_container_width=True)

        # Display the correlation values as a table for highly correlated and moderately correlated values
        st.write('Highly Correlated Values (>= 0.7):')
        highly_correlated = correlation_matrix[
            (correlation_matrix >= threshold_high) & (correlation_matrix < 1.0)
        ]
        st.write(highly_correlated)

        st.write('Moderately Correlated Values (>= 0.4):')
        moderately_correlated = correlation_matrix[
            (correlation_matrix >= threshold_moderate) & (correlation_matrix < threshold_high)
        ]
        st.write(moderately_correlated)


# Define a function to show distribution of total enrollment in a district by gender
def plot_distribution_by_gender(df):
    st.title('Distribution of Total Enrollment by Gender')
    district_options = ['Select'] + df['District'].unique().tolist()
    selected_district = st.selectbox('Select District to plot', district_options)
    if selected_district != 'Select':
        filtered_df = df[df['District'] == selected_district]

        fig = px.bar(
            data_frame=filtered_df,
            x='Gender',
            y='Total Enr',
            color='Gender',
            hover_name='District',
            hover_data=['School Name'],
            title=f'Distribution of Total Enr in {selected_district} by Gender',
            labels={'Total Enr': 'Total Enrollment'}
        )

        st.plotly_chart(fig)
def plot_heatmap(df):
    st.title('Correlation Heatmap for District')
    district_options = ['Select'] + df['District'].unique().tolist()
    selected_district = st.selectbox('Select District to show heatmap', district_options)
    if selected_district != 'Select':
        filtered_df = df[df['District'] == selected_district]
        numerical_columns = filtered_df.select_dtypes(include=['int64', 'float64'])

        fig = px.imshow(numerical_columns.corr(),
                        labels=dict(color='Correlation'),
                        title=f'Correlation Heatmap for {selected_district}',
                        color_continuous_scale='Viridis')

        fig.update_xaxes(side='top')
        fig.update_layout(height=1000, width=1000)
        fig.update_traces(hoverongaps=False, hovertemplate='%{y} and %{x}: Correlation: %{z:.2f}')

        st.plotly_chart(fig)


# Define a function to create a barplot for District and total number of schools
def plot_barplot_district_total_schools(df):
    if st.checkbox('Show Numerical Columns', key='show_numerical_columns_barplot'):
        st.title('Barplot for District & Total Number of Schools')
        data = df.groupby(['District']).size().rename('Total number of schools').reset_index()
        fig = px.bar(data, x='District', y='Total number of schools', color='District', text='Total number of schools')
        st.plotly_chart(fig)


# Define a function to compare a numerical column by district
def compare_numerical_column_by_district(df):
    st.title('Comparison of Numerical Column by District')
    numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
    # Exclude 'School Name' from the list of numerical columns
    numerical_columns = [col for col in numerical_columns if col != 'School Name']

    selected_numerical_column = st.selectbox("Select a Numerical Column to Compare", ['Select'] + list(numerical_columns))

    if selected_numerical_column != 'Select':
        data_grouped = df.groupby('District')[selected_numerical_column].mean().reset_index()

        fig = px.bar(data_grouped, x='District', y=selected_numerical_column, color='District',
                     labels={selected_numerical_column: selected_numerical_column},
                     title=f'Comparison of {selected_numerical_column} by District',
                     hover_data=[selected_numerical_column, data_grouped[selected_numerical_column].apply(lambda x: f'Mean: {x:.2f}')])

        #fig.update_traces(text=data_grouped[selected_numerical_column].apply(lambda x: f'Mean: {x:.2f}'),
         #                 textposition='outside')
        fig.update_layout(xaxis_title='District',
                          yaxis_title=selected_numerical_column, height=500)

        st.plotly_chart(fig)

# Define a function to compare a categorical column by district
def compare_categorical_column_by_district(df):
    st.title('Distribution of Categorical Column by District')

    # List of categorical columns in the DataFrame
    categorical_columns = df.select_dtypes(include='object').columns.tolist()

    selected_categorical_column = st.selectbox("Select a Categorical Column to Compare", ['Select'] + categorical_columns)

    # Check if a valid column is selected
    if selected_categorical_column != 'Select':
        data_grouped = df.groupby(['District', selected_categorical_column]).size().reset_index(name='Count')

        fig = px.bar(data_grouped, x='District', y='Count', color=selected_categorical_column,
                     labels={'Count': 'Number of Schools'},
                     title=f'Distribution of {selected_categorical_column} by District',
                     hover_data=[selected_categorical_column])

        fig.update_layout(xaxis_title='District', yaxis_title='Number of Schools')

        st.plotly_chart(fig)

# Define a function to display all sections
def display_all_sections(df):
    home()
    focus_on_enrollment()
    story()
    insights()
    show_raw_data(df)
    show_numerical_columns(df)
    show_categorical_columns(df)
    show_summary(df)
    plot_distribution_by_gender(df)
    plot_heatmap(df)
    plot_barplot_district_total_schools(df)
    compare_numerical_column_by_district(df)
    compare_categorical_column_by_district(df)
    show_correlation_heatmap_and_table(df)

# Sidebar navigation
def main():
    st.sidebar.title(' Menu')
    page = st.sidebar.selectbox("Select a Page", ["Home","Targeted column", "Story tellinng for project", "Data insights", "All Sections", "Raw Data", "Numerical Columns", "Categorical Columns",
                                                   "Summary Statistics", "Correlation Heatmap and Table", "Distribution by Gender", "Heatmap",
                                                   "Barplot for District", "Compare Numerical Column by District",
                                                   "Compare Categorical Column by District"])

    df = load_data()

    if page == "Home":
        home()
    elif page == "Targeted column":
        focus_on_enrollment()
    elif page == "Data insights":
        insights()
    elif page == "Story tellinng for project":
        story()
    elif page == "All Sections":
        display_all_sections(df)
    elif page == "Raw Data":
        show_raw_data(df)
    elif page == "Numerical Columns":
        show_numerical_columns(df)
    elif page == "Categorical Columns":
        show_categorical_columns(df)
    elif page == "Summary Statistics":
        show_summary(df)
    elif page == "Distribution by Gender":
        plot_distribution_by_gender(df)
    elif page == "Heatmap":
        plot_heatmap(df)
    elif page == "Barplot for District":
        plot_barplot_district_total_schools(df)
    elif page == "Compare Numerical Column by District":
        compare_numerical_column_by_district(df)
    elif page == "Compare Categorical Column by District":
        compare_categorical_column_by_district(df)
    elif page == "Correlation Heatmap and Table":
        show_correlation_heatmap_and_table(df)

if __name__ == "__main__":
    main()
```
